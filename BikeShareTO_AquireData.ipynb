```python
import pandas as pd
import numpy as np
import requests
import json
import sqlite3
import os
import time
```

```python
# Create sqlite database and cursor
conn = sqlite3.connect('bikeshareTO.db')
c = conn.cursor()
# Create the table of pitches
c.execute("""CREATE TABLE IF NOT EXISTS station_status (
            last_reported integer,
            station_id integer,
            num_bikes_available integer,
            num_bikes_disabled integer,
            num_docks_available integer,
            status text
            )""")
conn.commit()
```

```python
url = 'https://tor.publicbikesystem.net/ube/gbfs/v1/en/station_status'

while True:
    response = requests.get(url)
    df = pd.DataFrame(json.loads(response.text)['data']['stations'])[['last_reported','station_id','num_bikes_available','num_bikes_disabled','num_docks_available','status']]
    df.to_sql('station_status', conn, if_exists='append', index=False)
    time.sleep(60*10)
```

```python
pd.read_sql_query('SELECT * FROM station_status',conn)
```

```python
conn.close()
```
